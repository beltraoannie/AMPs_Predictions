```python
from collections import Counter

def mode(frame):
  col=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q',
        'R', 'S', 'T', 'V', 'W', 'Y']
  arr = []
  brr = []
  crr = []
  for i in col: 
    arr=frame[i].to_list()
    data = Counter(arr)
    brr =data.most_common(1)
    crr.append(brr[0][0])
  return crr
```

```python
def mean(frame):
  col=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q',
          'R', 'S', 'T', 'V', 'W', 'Y']
  soma = []
  tamanho = []
  media = []
  arr = []
  for i in col: 
    arr=frame[i].to_list()
    soma=sum(arr)
    tamanho=len(arr)
    media.append(soma/tamanho)
  return media
```

```python
def mode_plot(dataframe1, dataframe2):
  phytamp = mode(dataframe1)
  onekp= mode(dataframe2)
  df = pd.DataFrame({'PhytAMP': phytamp, '1KP': onekp})
  df.plot(kind='bar',cmap='RdYlBu')
  cols=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q',
          'R', 'S', 'T', 'V', 'W', 'Y']
  a= plt.xticks(np.arange(20),cols)
  a= plt.xlabel('Amino Acids')
  a= plt.ylabel('Mode')
  a= plt.legend(['PhytAMP', '1KP']);
  return a
```

```python
def mean_plot(frame1, frame2):
  phytamp = mean(frame1)
  onekp= mean(frame2)
  data = pd.DataFrame({'PhytAMP': phytamp, '1KP': onekp})
  data.plot(kind='bar',cmap='RdYlBu')
  cols=['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q',
          'R', 'S', 'T', 'V', 'W', 'Y']
  b= plt.xticks(np.arange(20),cols)
  b= plt.xlabel('Amino Acids')
  b= plt.ylabel('Mean')
  b= plt.legend(['PhytAMP', '1KP']);
  return b
```
